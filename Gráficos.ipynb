```python
#Importando as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Leitura e visualização dos dados
avc=pd.read_csv("Basebalanceada.csv")
```

```python
# Gerando gráfico
# AVC
plt.plot(['Menos de um ano', 'Um ano', 'Dois anos', 'Três anos'],
         [0, 262, 425, 288], color = 'red')

# Feminino - f
plt.plot(['Menos de um ano', 'Um ano', 'Dois anos', 'Três anos'],
         [0, 55, 88, 45], color = 'orange')

# Masculino - m
plt.plot(['Menos de um ano', 'Um ano', 'Dois anos', 'Três anos'],
         [0, 119,228, 156], color = 'blue')

plt.legend(['Total de Crianças', 'Sexo Feminino', 'Sexo Masculino'])
plt.xlabel('Faixas Etárias')
plt.ylabel('Número de Casos')
fig = plt.gcf()
plt.show()
fig.savefig('autismo-genero.png', format='png')
```

```python
#Fatores de risco 1
valores_sexo_M = [7,283,56,43]

valores_sexo_F = [8,170,95,50]

# Cria eixo x para produto A e produto B com uma separação de 0.25 entre as barras
x1 =  np.arange(len(valores_sexo_F))
x2 = [x + 0.25 for x in x1]

# Plota as barras
plt.bar(x1, valores_sexo_F, width=0.25, label = 'Sexo Feminino', color = 'b')
plt.bar(x2, valores_sexo_M, width=0.25, label = 'Sexo Masculino', color = 'orange')

# coloca o nome dos meses como label do eixo x
fator = ['Fibrilação atrial','Doença no pulmão', 'AVC na família', 'Transtorno nervoso']
plt.xticks([x + 0.25 for x in range(len(valores_sexo_F))], fator)

# inseri uma legenda no gráfico
plt.legend()

fig = plt.gcf()
fig.savefig('fatorrisco1.png', format='png')
plt.title("Casos de AVC")
plt.show()
```
